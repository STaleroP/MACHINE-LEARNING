```python
#Ciclos o bucles (while, for, break, continue)

#while condicion:
  #sentencias
print("Tabla con While")
i=0
while i<=10:
  print(f"5x{i}={5*i}")
  i=i+1 #o i+=1

print("\nTabla con for")
#Ciclo for
for i in range(0,11,1):
  print(f"5x{i}={5*i}")
```

```python
#Iterar con la función zip
lista_clase=["Juan", "Felipe", "Ana", "Julian"]
codigo_clase=["1001", "1002", "1003", "1004"]

a=zip(lista_clase, codigo_clase)
for i,j in a:
  print("Mi estudiante ", i, "y su codigo es:", j)
```
