# **Ciclo de Vida de un Proyecto en Minería de Datos**

El proceso de minería de datos sigue un ciclo estructurado que permite transformar datos crudos en información valiosa y conocimiento útil. A continuación, se describen sus principales etapas:

---

### 🔹 **1. Adquisición o Colección de Datos**
- Los datos pueden provenir de diferentes fuentes: bases de datos, archivos locales, APIs, sensores, web scraping, entre otros.  
- Los formatos más comunes incluyen: `.csv`, `.xlsx`, `.txt`, `.json`, `.dat`, entre otros.  
- También existen librerías y repositorios especializados (ej. *sklearn.datasets*, *UCI Machine Learning Repository*, *Kaggle*) que facilitan la obtención de conjuntos de datos de ejemplo.  

---

### 🔹 **2. Preparación de Datos**
- Es una de las fases más importantes, ya que los datos raramente vienen listos para su análisis.  
- Se incluyen tareas como:
  - **Limpieza:** manejo de valores nulos, duplicados, datos inconsistentes.  
  - **Transformación:** normalización, estandarización, creación de variables derivadas.  
  - **Codificación:** transformar variables categóricas en numéricas (ej. one-hot encoding).  
  - **Reducción de dimensionalidad:** eliminación de atributos irrelevantes o redundantes.  

---

### 🔹 **3. Proceso / Modelado**
- En esta fase se aplican los **algoritmos de aprendizaje automático** y técnicas estadísticas.  
- Dependiendo del objetivo, se pueden usar distintos enfoques:
  - **Clasificación** (ej. árboles de decisión, SVM, redes neuronales).  
  - **Regresión** (ej. regresión lineal, regresión logística).  
  - **Clustering** (ej. k-means, DBSCAN).  
  - **Reglas de asociación** (ej. algoritmo Apriori).  
- Aquí también se hace la **selección de modelos** y la **validación cruzada** para evaluar su desempeño.  

---

### 🔹 **4. Evaluación de Resultados**
- Se interpretan los resultados obtenidos y se contrastan con los objetivos iniciales.  
- Se utilizan métricas como:
  - Para **clasificación**: accuracy, precisión, recall, F1-score, matriz de confusión.  
  - Para **regresión**: error cuadrático medio (MSE), R², MAE.  
  - Para **clustering**: índice de silueta, Davies–Bouldin.  
- La evaluación asegura que el modelo sea confiable antes de desplegarlo.  

---

### 🔹 **5. Presentación y Visualización**
- Los resultados deben comunicarse de manera clara y efectiva.  
- Se emplean:
  - **Gráficos** (ej. histogramas, scatter plots, boxplots).  
  - **Tablas** con métricas comparativas.  
  - **Dashboards** interactivos para exploración visual.  
- La visualización permite descubrir patrones, tendencias y anomalías que ayudan en la toma de decisiones.  

---

### 🔹 **6. Implementación / Despliegue**
- Una vez validado, el modelo puede integrarse en sistemas de producción o aplicaciones reales.  
- Puede implicar:
  - Desarrollar una API para servir predicciones.  
  - Integración en un sistema de recomendación.  
  - Automatización de procesos analíticos.  
- Aquí también se incluye el **monitoreo y mantenimiento** del modelo para asegurar que siga siendo válido con nuevos datos.

---

# **Regresión para distintas rectas**

```python
# Cargar dataset iris
iris = load_iris(as_frame=True)
x = iris.frame[["petal length (cm)", "petal width (cm)"]]

# Rangos de pendientes (m) e interceptos (b)
pendientes = np.linspace(0.1, 1.5, 50)   # más fino que antes
interceptos = np.linspace(-1.0, 1.0, 50)

mejor_m, mejor_b, mejor_error = None, None, float("inf")

# Buscar la mejor recta
for m in pendientes:
    for b in interceptos:
        y_pred = m * x["petal length (cm)"] + b
        error = np.mean((x["petal width (cm)"] - y_pred) ** 2)  # MSE
        if error < mejor_error:
            mejor_error = error
            mejor_m, mejor_b = m, b

print(f"Mejor pendiente (m): {mejor_m:.4f}")
print(f"Mejor intercepto (b): {mejor_b:.4f}")
print(f"Error cuadrático medio (MSE): {mejor_error:.6f}")

# Graficar datos reales
plt.figure(figsize=(8,6))
plt.scatter(x["petal length (cm)"], x["petal width (cm)"], s=15, label="Datos reales")

# Graficar la mejor recta
y_pred_mejor = mejor_m * x["petal length (cm)"] + mejor_b
plt.plot(x["petal length (cm)"], y_pred_mejor, color="red", linewidth=2, label="Mejor recta")

plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend()
plt.show()
```
