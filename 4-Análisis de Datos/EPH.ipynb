¿Cuál era la tasa de desocupación en el período seleccionado
para las mujeres? ¿Era más alta o más baja que la de los
hombres? (Extra: pensar el resultado usando una función)

**Nota: la tasa se calcula como Desocupados / (Ocupados + Desocupados)**

### Tasa de desocupacion para los hombres

```python
eph_2.query('CH04==1 & ESTADO ==2').shape[0]/(eph_2.query('CH04==1 & ESTADO ==2').shape[0]+eph_2.query('CH04==1 & ESTADO ==1').shape[0])
```

### Tasa de desocupacion para las mujeres

```python
eph_2.query('CH04==2 & ESTADO ==2').shape[0]/(eph_2.query('CH04==2 & ESTADO ==2').shape[0]+eph_2.query('CH04==2 & ESTADO ==1').shape[0])
```

La tasa de **desocupacion** de los **hombres era del 7.47%**, mientras que de las **mujeres era del 9.14%***. Por lo tanto, las mujeres tenian una tasa mas alta de desocupacion

## Como funcion

```python
def calcular_tasa_desocupacion(df, genero):
    ocupados = df.query(f'CH04 == {genero} & ESTADO == 1').shape[0]
    desocupados = df.query(f'CH04 == {genero} & ESTADO == 2').shape[0]
    tasa_desocupacion = desocupados / (ocupados + desocupados)
    return tasa_desocupacion

# Calculando la tasa de desocupación para hombres (CH04 == 1)
tasa_hombres = calcular_tasa_desocupacion(eph_2, 1)
print("Tasa de desocupación para hombres:", tasa_hombres)

# Calculando la tasa de desocupación para mujeres (CH04 == 2)
tasa_mujeres = calcular_tasa_desocupacion(eph_2, 2)
print("Tasa de desocupación para mujeres:", tasa_mujeres)
```
