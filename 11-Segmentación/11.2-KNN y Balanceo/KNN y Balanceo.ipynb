```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
import seaborn as sb
%matplotlib inline
plt.rcParams["figure.figsize"] = (16, 9)
plt.style.use("ggplot")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
reviews_data = pd.read_csv("reviews_sentiment.csv", sep=";")
reviews_data.head()
```

```python
#Preprocess

X = reviews_data[["wordcount","sentimentValue"]].values
y= reviews_data["Star Rating"]
print(y.head())
scaler=MinMaxScaler()
X = scaler.fit_transform(X)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=137)

n_neighbors=[2,3,4,5,6,7,8,9,10]
for i in n_neighbors:
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train, y_train)
  print(f"Metrics for {i} neighbors")
  print(f"Acuracy (train, not that important):{knn.score(X_train,y_train)}")
  print(f"Acuracy (test):{knn.score(X_test,y_test)}")
  pred=knn.predict(X_test)
  print(classification_report(y_test,pred))
  print("===============================================")
```

```python
#Balanceo aquí mismito, oversamplinga hace lo mismo pero en español

from imblearn.over_sampling import RandomOverSampler

overthing = RandomOverSampler(random_state=42) #why 42? -no idea tbh
X_train_balanced, y_train_balanced =overthing.fit_resample(X_train, y_train) #this is a black box ish, we may need to check the code inside imblearn

print(y_train.value_counts())
print(y_train_balanced.value_counts()) # this generate data until reach higher feature value
```

```python
#X_train_o, X_test_o, y_train_o, y_test_o = train_test_split(X, y, test_size=0.3, random_state=137) this should be necessary tho

n_neighbors=[2,3,4,5,6,7,8,9,10]
for i in n_neighbors:
  print(f"Metrics for {i} neighbors")
  print(f"Acuracy (train, not that important):{knn.score(X_train_balanced,y_train_balanced)}")
  print(f"Acuracy (test):{knn.score(X_test,y_test)}")
  pred=knn.predict(X_test)
  print(classification_report(y_test,pred))
  print("===============================================")
```

```python

overthing = RandomOverSampler(random_state=42)
X_balanced, y_balanced = overthing.fit_resample(X, y)

print(y.value_counts())
print(y_balanced.value_counts())
```

```python
n_neighbors=[2,3,4,5,6,7,8,9,10]
for i in n_neighbors:
  print(f"Metrics for {i} neighbors")
  print(f"Acuracy (train, not that important):{knn.score(X_balanced,y_balanced)}")
  print(f"Acuracy (test):{knn.score(X_test,y_test)}")
  pred=knn.predict(X_test)
  print(classification_report(y_test,pred))
  print("===============================================")
```

```python
#undersampling stuff, less frequency
from imblearn.under_sampling import RandomUnderSampler

underthing = RandomUnderSampler(random_state=42)
X_under_train, y_under_train = underthing.fit_resample(X, y) #original indeed

print(y.value_counts())
print(y_under_train.value_counts())
```

```python
n_neighbors=[2,3,4,5,6,7,8,9,10]
for i in n_neighbors:
  print(f"Metrics for {i} neighbors")
  print(f"Acuracy (train, not that important):{knn.score(X_under_train,y_under_train)}")
  print(f"Acuracy (test):{knn.score(X_under_train,y_under_train)}")
  pred=knn.predict(X_test)
  print(classification_report(y_test,pred))
  print("===============================================")
```

```python
#existe una cosa llamada cross validation también

from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_val_predict #al parecer esto es una validación cruzada, se

cv = StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
cv_score = cross_val_score(KNeighborsClassifier(n_neighbors=3), X_train, y_train,cv=cv) #model, over original train data

print(f"Acc prom = {cv_score.mean()}") #not balanced cross val
```

```python
cv_b = StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
cv_score_b = cross_val_score(KNeighborsClassifier(n_neighbors=3), X_train_balanced, y_train_balanced,cv=cv_b) #model, over balanced train data
print(cv_b)
print(f"Acc prom = {cv_score_b.mean()}") #it grows better cuz there is more information
```

```python
cv_pred = cross_val_predict(KNeighborsClassifier(n_neighbors=3), X_test, y_test, cv=cv_b) #this is why it is important to re-divide the data before, in order to predict over other balanced data
print(cv_pred)
```
