# **Método de Regresión Lineal**

El modelo de **regresión lineal** representa un marco metodológico que permite analizar la relación entre dos variables, asumiendo una hipótesis de causalidad entre ellas.  

Se define una hipótesis teórica representada por medio de una función matemática:

\begin{equation*}
y = f(x)
\end{equation*}

donde la variable $y$ es una función de la variable $x$. Esta función puede ser lineal o no lineal.  

En el caso lineal, con dos variables $(y, x)$ tomadas sobre la misma población, se construye un **diagrama de dispersión**:  
- En el eje *x* se ubica la variable independiente.  
- En el eje *y* se ubica la variable dependiente.  

Si las variables están correlacionadas, el gráfico muestra cierta tendencia; si no, los puntos estarán dispersos al azar.

---

## **Modelo estadístico general**

Una primera aproximación es el modelo de regresión lineal simple:

\begin{equation*}
y_i = \alpha + \beta x_i \qquad i = 1,2,...,n
\end{equation*}

donde:  
- $y$ = variable dependiente  
- $x$ = variable independiente o explicativa  
- $\alpha$ y $\beta$ = parámetros del modelo  

El modelo establece que un cambio de una unidad en $x$ produce un cambio en $y$, medido por $\beta$.  


El modelo establece que un cambio en una unidad de $x$ produce u ocasiona un cambio en la variable $y$, medido por el parámetro $\beta$. Sin embargo, a manera de ejemplo, una media poblacional de $y$ es una característica desconocida de la distribución, por lo tanto la pendiente es desconocida. Así el supuesto básico del modelo de regresión es que una realización muestral de la variable $y$ puede ser expresada como una combinación lineal de las observaciones de $x$ incluyendo un **término de error $u_i$**:

\begin{equation*}
y_i = \alpha + \beta x_i + u_i \qquad i = 1,2,...,n
\end{equation*}

---

Existe una diferencia entre los datos del experimento y los valores predichos por la recta del modelo planteado. Esa diferencia se denota el error $u_i$; las desviaciones en cada observación generan una serie de errores, la cual se considera como una variable aleatoria alrededor del cero, con valores positivos y negativos.

Para cada valor de $x$ existe una distribución de probabilidad del término de error y por consiguiente una distribución de probabilidad para la variable $y$. Asumiendo que el término de error presenta una distribución de probabilidad se realizan entonces ciertos supuestos sobre dicha distribución:

1. **Valor esperado del error nulo $E(u_i) = 0$:** El término aleatorio tiene esperanza igual a cero para todas las observaciones. Este supuesto implica que en promedio la relación entre $y$ y $x$ es exactamente lineal, aunque las realizaciones particulares de los $u_i$'s pueden ser distintas de cero.
2. **Homoscedasticidad $Var(u_i) = \sigma^2$:** Varianza constante para todas las observaciones.
3. **Normalidad:** los errores siguen una distribución normal, con media cero y varianza constante.
4. **Ausencia de autocorrelación $Cov(u_i,u_j) = 0$, $\forall i \neq j$:** Las covarianzas del término aleatorio entre dos observaciones distintas son iguales a cero.
5. **Forma funcional lineal:** La relación entre $x$ y $y$ es lineal.  
6. **Parámetros constantes:** $\alpha$ y $\beta$ no cambian a lo largo de la muestra.  
7. **Media condicional nula del error $E(u_i|x_i) = 0$:** Este supuesto indica que los factores que no están incluidos en el modelo, están incorporados en el término de error, no afectan sistemáticamente el valor de la media de $y$.

Estos supuestos permiten construir la base teórica para estimar parámetros y hacer inferencia estadística.

## **Método de estimación de Mínimos Cuadrados Ordinarios (MCO)**

La forma de obtener una estimación de los parámetros del modelo previamente planteado para el ajuste de datos es por medio del **MÉTODO DE MÍNIMOS CUADRADOS ORDINARIOS (MCO)**. Nótese que para cada punto hay una distancia que será igual a la diferencia del valor real $y$ entre el valor predicho por el modelo $\hat{y}=\alpha+\beta x_i$. Definimos entonces:
\begin{equation*}
\varepsilon_n^2=(y_n-\hat{y}_n)^2
\end{equation*}

Dicha función debe ser minimizada. Si definimos la distancia como el término de error al cuadrado, y la distancia total como la suma de los errores al cuadrado (SEC), tenemos del modelo lineal con error:

\begin{equation*}
\sum_{i=1}^N u_i^2=\sum_{i=1}^N\left(y_i-\alpha-\beta x_i\right)^2
\end{equation*}

El método de mínimos cuadrados ordinarios (MCO) considera los valores muestrales de los parámetros, los cuales se definen como los estimadores de MCO que deben ser calculados al minimzar el error . Esto es un problema encasillado en la optimización, derivando respecto a los parámetros $\alpha$ y $\beta$ e igualando a cero.

\begin{equation*}
S(\hat{u})_{min}=\sum_{i=1}^N\left(y_i-\alpha-\beta x_i\right)^2
\end{equation*}

Luego,

\begin{equation*}
\begin{aligned}
& \frac{\partial S(\hat{u})}{\partial \alpha}=2 \sum_{i=1}^N\left(y_i-\alpha-\beta x_i\right)(-1)=0 \\
& \frac{\partial S(\hat{u})}{\partial \beta}=2 \sum_{i=1}^N\left(y_i-\alpha-\beta x_i\right)\left(-x_i\right)=0
\end{aligned}
\end{equation*}

Desarrollando términos

\begin{equation*}
\begin{aligned}
& \frac{\partial S(\hat{u})}{\partial \alpha}=-2 \sum_{i=1}^N y_i+2 \sum_{i=1}^N \alpha+2 \sum_{i=1}^N \beta x_i=0 \\
& \frac{\partial S(\hat{u})}{\partial \beta}=-2 \sum_{i=1}^N y_i x_i+2 \sum_{i=1}^N \alpha x_i+2 \sum_{i=1}^N \beta x_i^2=0
\end{aligned}
\end{equation*}

Esto es un sistema de dos ecuaciones con dos incógnitas. Despejando $\alpha$ de la primera ecuación se obtiene el siguiente resultado:

$$ \alpha=\overline{y}-\beta \overline{x} $$

siendo $\overline{y}$ la suma de los $N$ datos de $y$ junto con $\overline{x}$ siendo la suma de los $N$ datos de $x$. Luego reemplazamos $\alpha$ en la segunda ecuación:

$$
\begin{gathered}
-2 \sum_{i=1}^N y_i x_i+2 \sum_{i=1}^N(\bar{y}-\beta x) x_i+2 \sum_{i=1}^N \beta x_i^2=0 \\
-\sum_{i=1}^N y_i x_i+\sum_{i=1}^N(\bar{y}-\beta \bar{x}) x_i+\sum_{i=1}^N \beta x_i^2=0
\end{gathered}
$$

Reordenando la ecuación
$$
-\sum_{i=1}^N y_i x_i+\sum_{i=1}^N \bar{y} x_i-\beta \sum_{i=1}^N \bar{x} x_i+\sum_{i=1}^N \beta x_i^2=0
$$

Agrupando términos semejantes:
$$
-\sum_{i=1}^N y_i x_i+y \sum_{i=1}^N x_i+\beta\left(\sum_{i=1}^N x_i^2-x \sum_{i=1}^N x_i\right)=0
$$

El estimador $\beta$ se puede calcular como:
$$
\beta=\dfrac{\sum_{i=1}^N y_i x_i-\bar{y} \sum_{i=1}^N x_i}{\left(\sum_{i=1}^N x_i^2-\bar{x} \sum_{i=1}^N x_i\right)}=\dfrac{\sum_{i=1}^N\left(y_i-\bar{y}\right)\left(x_i-\bar{x}\right)}{\sum_{i=1}^N\left(x_i-\bar{x}\right)^2}=\frac{\operatorname{Cov}(y x)}{\operatorname{Var}(x)}
$$

El modelo de regresión lineal establece que $y_i$ puede ser aproximada por una función lineal de $x_i$, y el término de error $u_i$ es la discrepancia entre la variable
observada y el modelo. Los valores de $\alpha$ y $\beta$ permiten “proyectar” o “estimar” los valores de la variable dependiente a partir de valores de la variable explicativa.

```python
import matplotlib.pyplot as plt
import numpy as np

# Función para generar datos con ruido
def genera_datos(n_puntos, pendiente, corte, ds, seed=None):
    if seed is not None:
        np.random.seed(seed)  # Semilla opcional
    x = np.linspace(0, 10, n_puntos)
    y = corte + pendiente * x + np.random.normal(0, ds, n_puntos)
    return x, y

# Parámetros
n_max = 800
incremento_n = 500
ds_max = 6
incremento_ds = 1

for n_puntos in range(incremento_n, n_max + incremento_n, incremento_n):
    for ds in range(1, ds_max + incremento_ds, incremento_ds):
        x, y = genera_datos(n_puntos, 2.8, 5, ds, seed=0)

        # Ajuste de regresión lineal (pendiente y corte)
        beta, alfa = np.polyfit(x, y, 1)

        # Graficar solo algunos casos para no saturar
        if ds in [1, ds_max] and n_puntos in [incremento_n, n_max]:
            plt.scatter(x, y, label="Datos", alpha=0.5)
            plt.plot(x, 5 + 2.8 * x, "r", label="Recta usada")
            plt.plot(x, alfa + beta * x, "g--", label="Recta aprox")
            plt.xlabel("Datos X")
            plt.ylabel("Datos Y")
            plt.title(f"n_puntos = {n_puntos}, ds = {ds}")
            plt.legend()
            plt.show()

        print(f"n_puntos={n_puntos}, ds={ds} → beta={beta:.3f}, alfa={alfa:.3f}")
```

```python
#Replicamos la funcion creada en clase
def genera_datos(n_puntos, pendiente, corte, ds):
    np.random.seed(0)  #Semilla para generar los mismos
    x = np.linspace(0, 10, n_puntos)
    y = pendiente * x + corte + np.random.normal(0, ds, n_puntos)
    return x, y

#Creamos un ciclo para añadir datos y aumentar ds por cada aumento de datos
n_max = 1000 #Con 1'000,000 explotó :'v
incremento_n = 500
ds_max = 6
incremento_ds = 1

for n_puntos in range(incremento_n, n_max + incremento_n, incremento_n):
    for ds in range(1, ds_max + incremento_ds, incremento_ds):
        x, y = genera_datos(n_puntos, 2.8, 5, ds)
        #Calculo de los parámetros
        beta = (n_puntos * np.sum(x * y) - np.sum(x) * np.sum(y)) / (n_puntos * np.sum(x * x) - np.sum(x) ** 2)
        alfa = (np.sum(y) * np.sum(x * x) - np.sum(x) * np.sum(x * y)) / (n_puntos * np.sum(x * x) - np.sum(x) ** 2)

        plt.scatter(x, y, label="Datos")
        plt.plot(x, 5 + 2.8 * x, color="red", label="Recta usada")
        plt.plot(x, alfa + beta * x, color="green", label="Recta aprox")
        plt.xlabel("Datos X")
        plt.ylabel("Datos Y")
        plt.title(f"n_puntos = {n_puntos}, ds = {ds}")
        plt.legend()
        plt.show()

        print(beta)
        print(alfa)
```
