# Regresión usando `LinearRegression` y `SGDRegressor`

## Importación de las librerías

```python
import pandas as pd
import timeit
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDRegressor
```

### Regresión lineal para el DataFrame sin escalar

Importación del DataFrame general.

```python
df = pd.read_excel("DatosRegresion15D.xlsx")
df
```

Selección de las variables `X` y `Y`.

```python
X = df.drop("consumo",axis=1)
Y = df['consumo']
```

Selección de las variables de entrenamiento y de evaluación.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20)
```

Aplicación de los métodos `linear_model.LinearRegression()` y `SGDRegressor()`.

```python
regr = linear_model.LinearRegression()
sgd = SGDRegressor(max_iter = 150, tol = 1e-3)
```

Detección del tiempo que tarda en entrenar el método `linear_model.LinearRegression()`.

```python
start_time = timeit.default_timer()
###
regr.fit(X_train, Y_train)
###
end_time = timeit.default_timer()

elapsed_time_ps = (end_time - start_time) * 1e12
print(f'Tiempo transcurrido: {elapsed_time_ps:.3f} ps')
```

Impresión de los coeficientes y de las métricas.

```python
print(regr.coef_)
Y_prec_regr = regr.predict(X_test)

metrica_msr_regr = mean_squared_error(Y_test, Y_prec_regr)
print(metrica_msr_regr)
metrica_r2_regr = r2_score(Y_test, Y_prec_regr)
print(metrica_r2_regr)
```

Detección del tiempo que tarda en entrenar el método `SGDRegressor()`.

```python
start_time = timeit.default_timer()
###
sgd.fit(X_train, Y_train)
###
end_time = timeit.default_timer()

elapsed_time_ps = (end_time - start_time) * 1e12
print(f'Tiempo transcurrido: {elapsed_time_ps:.3f} ps')
```

Impresión de los coeficientes y las métricas.

```python
print(sgd.coef_)
Y_prec_sgd = sgd.predict(X_test)

metrica_msr_sgd = mean_squared_error(Y_test, Y_prec_sgd)
print(metrica_msr_sgd)
metrica_r2_sgd = r2_score(Y_test, Y_prec_sgd)
print(metrica_r2_sgd)
```

A continuación se repiten todos los pasos pero escalando de diferentes maneras, estas son: media varianza, max_min y media nula.
