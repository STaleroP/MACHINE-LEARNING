¿Cuántas mujeres hay en la base de datos? ¿Y hombres?

```python
mujeres = eph_2.query('CH04 == 1')
mujeres.shape[0]
```

```python
hombres = eph_2.query('CH04 == 2')
hombres.shape[0]
```

¿Cuál era la tasa de desocupación en el período seleccionado para las mujeres? ¿Era más alta o más baja que la de los hombres?

```python
eph_ocupadas = mujeres.query('ESTADO == 1')
eph_desocupadas = mujeres.query('ESTADO == 2')

tasa = eph_ocupadas.shape[0]/(eph_ocupadas.shape[0]+eph_desocupadas.shape[0])
print(tasa)
```

```python
eph_ocupados = hombres.query('ESTADO == 1')
eph_desocupados = hombres.query('ESTADO == 2')

tasa = eph_ocupados.shape[0]/(eph_ocupados.shape[0]+eph_desocupados.shape[0])
print(tasa)
```

```python
def calcular_tasa_ocupacion(data):
    eph_ocupados = data.query('ESTADO == 1')
    eph_desocupados = data.query('ESTADO == 2')

    tasa_ocupacion = eph_ocupados.shape[0] / (eph_ocupados.shape[0] + eph_desocupados.shape[0])
    return tasa_ocupacion

tasa = calcular_tasa_ocupacion(mujeres)
print(tasa)
tasah = calcular_tasa_ocupacion(hombres)
print(tasah)
```
