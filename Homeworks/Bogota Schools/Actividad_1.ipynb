```python
df_propiedad_Planta_Fisica = df.query('propiedad_Planta_Fisica == "PERSONA NATURAL" | propiedad_Planta_Fisica == "COMUNIDAD RELIGIOSA"')
print("Los colegios donde la propiedad de la planta física es de una persona natural o de una comunidad religiosa se pueden ver en la siguiente tabla:\n")
df_propiedad_Planta_Fisica
```

```python
df_capacidades_Excepcionales = df_propiedad_Planta_Fisica['capacidades_Excepcionales'].notna()

if df_capacidades_Excepcionales.any():
    if df_capacidades_Excepcionales.sum() == df_propiedad_Planta_Fisica.shape[0]:
      print("Sí, ¡todos los colegios son para capacidades excepcionales!")
      df_capacidades_Excepcionales = pd.DataFrame()
    else:
      print("Sí, hay un total de",df_capacidades_Excepcionales.sum(),"colegios para capacidades especiales y estos son:\n")
      df_capacidades_Excepcionales = df_propiedad_Planta_Fisica.loc[df_capacidades_Excepcionales,:]
else:
    print("No hay colegios que sean para capacidades excepcionales.")
    df_capacidades_Excepcionales = pd.DataFrame()

pd.set_option('display.max_columns', None)
df_capacidades_Excepcionales if not df_capacidades_Excepcionales.empty else None
```

```python
df['niveles'] = df['niveles'].astype(str)

# Divide las cadenas en cada celda por comas y cuenta la longitud de la lista resultante
df['max_palabras'] = df['niveles'].str.split(',').apply(len)

# Encuentra la fila con el máximo número de palabras
df_niveles = df[df['max_palabras'] == df['max_palabras'].max()]

print("Los colegios que ofrecen todos los niveles educativos son:\n")
# Imprime la fila con el máximo número de palabras
df_niveles.drop('max_palabras', axis=1)
```
