### ¿Cuáles son los colegios que ofrecen todos los niveles educativos y qué idiomas ofrecen?

Para conocer cuáles son los colegios que ofrecen todos los niveles educativos se parte de una suposición la cual es: *"los colegios que, en principio, ofrecen todos los niveles educativos deben ser aquellos que ofrecen más niveles educativos en comparación al resto de la base de datos"*. Esta suposición se hace debido a que fuera de la base de datos, no se tiene ninguna otra información de cuáles son todos los niveles educativos para poder crear un filtro correspondiente.

Haciendo caso a la suposición, para conocer el número máximo de palabras en la columna "niveles" se cambia el tipo de dato de la columna por `str` o "string" (texto) usando el método `.astype()`, luego con ayuda del método `.str.split()` el cual divide el contenido de cada celda de la columna "niveles" en una lista usando como separador identificador "," y con ayuda del método `.apply()` el cual aplica la misma función a cada elemento de una serie, que para este caso la función sería `len` la cual nos daría la longitud de cada lista generada por `.str.split()` para cada fila, se obtiene una nueva columna llamada "max_palabras" en donde se guarda la cantidad de palabras de cada fila para la columna "niveles".

Posteriormente, reescribiendo en un nuevo DataFrame llamado `df_niveles` aquellas filas que cumplen la condición `df['max_palabras'] == df['max_palabras'].max()` (el comando `.max()` sirve para conocer el valor más grande en una serie), se imprime dicho DataFrame eliminando la nueva columna creada usando el comando `.drop()`.

Ya que se muestran todas las columnas del DataFrame original solo que filtrado para aquellos colegios que ofrecen todos los niveles educativos, **ahí mismo se puede consultar cuáles son los idiomas que ofrecen**.

```python
df['niveles'] = df['niveles'].astype(str)

df['max_palabras'] = df['niveles'].str.split(',').apply(len)

df_niveles = df[df['max_palabras'] == df['max_palabras'].max()]

print("Los colegios que ofrecen todos los niveles educativos son:\n")

df_niveles.drop('max_palabras', axis=1)
```

### ¿Cuál es la tasa de desocupados en el periodo seleccionado para las mujeres?, ¿es más alta o más baja que la de los hombres?

Para calcular la tasa de "desocupados" en general, se crea una función llamada `tasa_desocupados` la cual pide los valores de dos variables, `desocupados` y `ocupados`, para luego operarlas y obtener la tasa de "desocupados".

```python
def tasa_desocupados(desocupados,ocupados):
    return desocupados / (ocupados + desocupados)
```

Así, guardando la cantidad de "desocupados" y "ocupados" para hombres y mujeres de cada base de datos en una variable independiente para cada uno (todo esto con ayuda de los métodos ya vistos `.query()` y `.ESTADO.count()`) se calcula la tasa para hombres y mujeres de cada base de datos usando la función `tasa_desocupados` ya creada y se compara con un sencillo condicional. Finalmente se imprime en pantalla los resultados obtenidos.

```python
m21_desocupado = eph_2_21anos.query('SEXO and ESTADO == 2').ESTADO.count()
m21_ocupado = eph_2_21anos.query('SEXO == 2 and ESTADO == 1').ESTADO.count()
h21_desocupado = eph_2_21anos.query('SEXO == 1 and ESTADO == 2').ESTADO.count()
h21_ocupado = eph_2_21anos.query('SEXO and ESTADO == 1').ESTADO.count()

m40_desocupado = eph_2_40anos.query('SEXO and ESTADO == 2').ESTADO.count()
m40_ocupado = eph_2_40anos.query('SEXO == 2 and ESTADO == 1').ESTADO.count()
h40_desocupado = eph_2_40anos.query('SEXO == 1 and ESTADO == 2').ESTADO.count()
h40_ocupado = eph_2_40anos.query('SEXO and ESTADO == 1').ESTADO.count()

if tasa_desocupados(m21_desocupado,m21_ocupado) > tasa_desocupados(h21_desocupado,h21_ocupado):
    print("La tasa de desocupados de las mujeres, la cual es",round(tasa_desocupados(m21_desocupado,m21_ocupado),5),"aproximadamente, es más alta que la de los hombres para la primera base de datos.")
else:
    print("La tasa de desocupados de las mujeres, la cual es",round(tasa_desocupados(m21_desocupado,m21_ocupado),5),"aproximadamente, es más baja que la de los hombres para la primera base de datos.")
print()
if tasa_desocupados(m40_desocupado,m40_ocupado) > tasa_desocupados(h40_desocupado,h40_ocupado):
    print("La tasa de desocupados de las mujeres, la cual es",round(tasa_desocupados(m40_desocupado,m40_ocupado),5),"aproximadamente, es más alta que la de los hombres para la segunda base de datos.")
else:
    print("La tasa de desocupados de las mujeres, la cual es",round(tasa_desocupados(m40_desocupado,m40_ocupado),5),"aproximadamente, es más baja que la de los hombres para la segunda base de datos.")
```
