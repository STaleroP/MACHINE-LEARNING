```python
import numpy as np
import matplotlib.pyplot as plt

#Creamos los valores de theta
theta_values = np.arange(-np.pi, 2*np.pi + np.pi/4, np.pi/4)

#Calculamos los valores de sin(theta) y sin(theta + pi/2)
sin_theta_values = np.sin(theta_values)
sin_theta_plus_pi2_values = np.sin(theta_values + np.pi/2)

#Calculamos la suma de sin(theta) y sin(theta + pi/2)
sum_values = sin_theta_values + sin_theta_plus_pi2_values

#Creamos la tabla
table = np.column_stack((theta_values, sin_theta_values, sin_theta_plus_pi2_values, sum_values))
headers = ["theta", "sin(theta)", "sin(theta + pi/2)", "sin(theta) + sin(theta + pi/2)"]

#Imprimimos la tabla
print("{:<12} {:<12} {:<19} {:<30}".format(*headers))
print("-" * 75)
for row in table:
    print("{:<12.4f} {:<12.4f} {:<19.4f} {:<30.4f}".format(*row))

#Graficamos las funciones
plt.figure(figsize=(10, 6))
plt.plot(theta_values, sin_theta_values, label='sin(theta)')
plt.plot(theta_values, sin_theta_plus_pi2_values, label='sin(theta + pi/2)')
plt.plot(theta_values, sum_values, label='sin(theta) + sin(theta + pi/2)')
plt.title('Plot of sin(theta), sin(theta + pi/2), and sin(theta) + sin(theta + pi/2)')
plt.xlabel('theta')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

## We smooth the functions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

#Creamos los valores de theta
theta_values = np.arange(-np.pi, 2*np.pi + np.pi/4, np.pi/4)

#Calculamos los valores de sin(theta) y sin(theta + pi/2)
sin_theta_values = np.sin(theta_values)
sin_theta_plus_pi2_values = np.sin(theta_values + np.pi/2)

#Calculamos la suma de sin(theta) y sin(theta + pi/2)
sum_values = sin_theta_values + sin_theta_plus_pi2_values

#Interpolamos los datos para suavizar las gráficas
theta_interp = np.linspace(-np.pi, 2*np.pi, 1000)
sin_theta_interp = interp1d(theta_values, sin_theta_values, kind='cubic')(theta_interp)
sin_theta_plus_pi2_interp = interp1d(theta_values, sin_theta_plus_pi2_values, kind='cubic')(theta_interp)
sum_interp = interp1d(theta_values, sum_values, kind='cubic')(theta_interp)

#Graficamos las funciones suavizadas
plt.figure(figsize=(10, 6))
plt.plot(theta_interp, sin_theta_interp, label='sin(theta)')
plt.plot(theta_interp, sin_theta_plus_pi2_interp, label='sin(theta + pi/2)')
plt.plot(theta_interp, sum_interp, label='sin(theta) + sin(theta + pi/2)')
plt.title('Plot of sin(theta), sin(theta + pi/2), and sin(theta) + sin(theta + pi/2)')
plt.xlabel('theta')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

## Observations

1. The function $\sin \theta$ oscillates between -1 and 1 over the interval $
[-\pi, 2\pi]$.
2. The function $\sin \left(\theta + \frac{\pi}{2}\right)$ is essentially the same as $\cos \theta$, which is the sine function shifted to the right by $\frac{\pi}{2}$. It oscillates between -1 and 1 as well but starts at 1 where $\sin \theta$ is 0.
3. The function $\sin \theta + \sin \left(\theta + \frac{\pi}{2}\right)$ is the sum of the two functions above and is symmetric about $\theta = \pi$.
4. The function $\sin \theta + \sin \left(\theta + \frac{\pi}{2}\right)$ has a maximum magnitude when both $\sin \theta$ and $\sin \left(\theta + \frac{\pi}{2}\right)$ are either maximum or minimum, resulting in constructive interference.
5. The function $\sin \theta + \sin \left(\theta + \frac{\pi}{2}\right)$ has a minimum magnitude when one of $\sin \theta$ or $\sin \left(\theta + \frac{\pi}{2}\right)$ is maximum while the other is minimum, resulting in destructive interference.
