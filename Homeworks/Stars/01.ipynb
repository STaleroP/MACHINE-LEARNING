# *Pivot Tables (Tablas Dinámicas)*

***Star dataset to predict star types***

A 6 class star dataset for star classification with Deep Learned approaches

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from IPython.display import Image
from IPython.core.display import HTML

plt.style.use('ggplot')
%matplotlib inline
```

```python
# Leemos nuestra base de datos del Drive
solecito = pd.read_csv("stars.csv")
solecito
```

```python
#Mostramos el radio promedio para una combinación de clase espectral y color de estrella.
solecito_panel = pd.pivot_table(
    solecito,
    values= "Radius(R/Ro)",
    index= "Spectral Class",
    columns= "Star color",
    aggfunc="mean"
)
solecito_panel
```

```python
solecito_panel.columns
```

```python
#Graficamos algunas columnas de solecito_panel con un box plot para ver el promedio de radio de las estrellas
plt.title('Promedio de radio (R/Ro) por Color de Estrella')
plt.xlabel('Color de Estrella')
plt.ylabel('Promedio de radio (R/Ro)')
boxplot = solecito_panel.boxplot(column=['Red','Blue White', 'Blue', 'Blue-White', 'Orange', 'Orange-Red', 'White'])
```

```python
#Mostramos la temperatura promedio para cada combinación de tipo de estrella y color de estrella.
solecito_2 = pd.pivot_table(
    solecito,
    values='Temperature (K)',
    columns='Star color',
    index='Star type',
    aggfunc='mean'
    )
solecito_2
```

```python
#Graficamos algunas columnas de solecito_2 con un box plot para ver el promedio de temperatura de las estrellas
plt.title('Promedio de temperatura (k) según el color de Estrella')
plt.xlabel('Color de Estrella')
plt.ylabel('Promedio de temperatura (k)')
boxplot = solecito_2.boxplot(column=['Red','Blue White', 'Blue', 'Blue-White', 'Orange', 'Orange-Red', 'White'])
```

```python
#Mostramos la luminosidad promedio para cada tipo de estrella en función de su clase espectral
solecito_3 = pd.pivot_table(solecito,
                                  values='Luminosity(L/Lo)',
                                  columns='Star type',
                                  index='Spectral Class',
                                  aggfunc='mean')
solecito_3
```

```python
#Graficamos solecito_3
solecito_3.plot(kind='bar')
plt.xlabel('Clase espectral')
plt.ylabel('Promedio de la luminosidad (L/Lo)')
plt.title('Luminosidad para el tipo de estrella y su clase espectral')
plt.show()
```

```python
#Mostramos la magnitud absoluta promedio para cada tipo de estrella en función del color de la estrella
solecito_4 = pd.pivot_table(solecito, values='Absolute magnitude(Mv)', columns='Star type', index='Star color', aggfunc='mean')
solecito_4
```

```python
solecito_4.columns
```

```python
#Graficamos solecito_4
solecito_4.plot(kind='bar')
plt.xlabel('Color estrella')
plt.ylabel('Promedio magnitud absoluta (Mv)')
plt.title('Magnitud absoluta para el tipo de estrella y color')
plt.show()
```
