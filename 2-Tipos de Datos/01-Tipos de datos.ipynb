# **¿Cómo describimos un dato?**

Los datos son la materia prima de cualquier análisis. Para comprenderlos, es necesario clasificarlos según su naturaleza y la forma en que se organizan.

---

### 🔹 **Tipos de datos**
1. **Cualitativos (o categóricos):**
   - Representan información **descriptiva**, no numérica.
   - Ejemplo: *"Tomo café todos los días"*, *"Color de ojos: verde"*, *"Categoría: Aprobado/Reprobado"*.  
   - Pueden subdividirse en:
     - **Nominales:** no tienen orden natural (ej. colores, nombres, géneros).  
     - **Ordinales:** tienen un orden implícito (ej. nivel educativo: primaria, secundaria, universidad).  

2. **Cuantitativos (o numéricos):**
   - Representan información en forma de números.
   - Ejemplo: *Edad = 25*, *Peso = 65.3 kg*.  
   - Pueden subdividirse en:
     - **Discretos:** toman valores enteros contables (ej. número de hijos, cantidad de autos).  
     - **Continuos:** pueden tomar cualquier valor en un rango (ej. altura, temperatura, tiempo).  

---

### 🔹 **Estructura de los datos en análisis**
La estructura más usada en Python para trabajar con datos tabulares es el **DataFrame** (de la librería *pandas*).

- **Columnas:** representan **atributos, variables, características o dimensiones** del dataset.  
- **Filas:** representan **objetos, instancias o registros** del dataset.  

Ejemplo en Python:

```python
import pandas as pd

data = {
    "Nombre": ["Ana", "Luis", "Carla"],
    "Edad": [23, 31, 29],
    "Café_diario": ["Sí", "No", "Sí"]
}

df = pd.DataFrame(data)
print(df)


# **Atributos y Métricas**

En minería de datos, es fundamental entender qué tipo de atributos tenemos y bajo qué **escala de medición** se encuentran, ya que de ello depende el tipo de análisis y operaciones que se pueden realizar.  

---

### 🔹 **Atributo**
- Propiedad o característica que puede variar de un objeto a otro.  
- Puede ser **cualitativo** o **cuantitativo**.  

### 🔹 **Escala de medición o métrica**
- Define cómo se mide el atributo y qué operaciones son válidas.  
- Ejemplo: centímetros, metros, grados Celsius, Kelvins.  
- Permite hacer comparaciones y cálculos coherentes.  

---

## **Tipos de Atributos**

### 1. **Cualitativo o *categórico***
- Representan categorías o etiquetas, no valores numéricos.  
- Se tratan como **símbolos**.  
- Pueden ser:  
  - **Nominales:** solo permiten comprobar igualdad o diferencia (`==, !=`).  
    - Ejemplo: nombre, raza, marca, color, ID, tipo de bacterias.  
  - **Ordinales:** además de igualdad, permiten ordenar (`==, !=, <, >`).  
    - Ejemplo: rankings, notas, niveles (bajo, medio, alto).  

📌 **Operaciones válidas:** igualdad, desigualdad, orden (solo en ordinales).  

---

### 2. **Cuantitativo o *numérico***
- Representan cantidades numéricas.  
- Se tratan como **números** reales.  
- Pueden ser:  
  - **De Intervalo:** tienen cero arbitrario, se puede medir distancia, pero no proporción.  
    - Ejemplo: fechas en un calendario, temperaturas en °C o °F.  
  - **De Razón:** tienen cero absoluto (ausencia total de la magnitud), permiten todas las operaciones matemáticas y proporciones.  
    - Ejemplo: temperatura en Kelvin, peso, altura, duración en horas.  

📌 **Operaciones válidas:** igualdad, desigualdad, suma, resta (intervalo), multiplicación y división (razón).  

---

## **Discretos vs Continuos**
- **Discretos:** valores finitos o contables.  
  - Ejemplo: número de hijos, número de autos, variable binaria (0/1).  
- **Continuos:** pueden tomar cualquier valor dentro de un rango.  
  - Ejemplo: altura, peso, tiempo, velocidad.  


# **Formato de Datos**

En minería de datos, la forma en que se almacenan y organizan los datos es crucial para su análisis. Existen tres grandes categorías principales y algunos subtipos relevantes:

---

### 🔹 **1. Datos Estructurados**
- Datos con un **formato fijo y predefinido**.  
- Se organizan en tablas (filas = registros, columnas = atributos).  
- Fácilmente almacenados en bases de datos relacionales.  
- **Ejemplos de formatos:** `.csv`, `.xlsx`, `.sql`.  

📌 **Ejemplo en tabla:**

| ID | Nombre | Edad | Ciudad    |
|----|--------|------|-----------|
| 1  | Ana    | 23   | Bogotá    |
| 2  | Luis   | 31   | Medellín  |

---

### 🔹 **2. Datos Semi-Estructurados**
- No siguen una estructura tabular rígida.  
- Se organizan en forma jerárquica (nodos, etiquetas, atributos).  
- Utilizan *tags* o pares clave–valor para representar información.  
- **Ejemplos de formatos:** `.json`, `.xml`, `.yaml`.  

📌 **Ejemplo en JSON:**

```json
{
  "ID": 1,
  "Nombre": "Ana",
  "Edad": 23,
  "Ciudad": "Bogotá"
}
```

---

## 🔹 **3. Datos No Estructurados**  

- Carecen de un formato definido o predecible.  
- Incluyen **texto libre, imágenes, audio, video, señales**.  
- Requieren técnicas avanzadas de **minería de texto, NLP o visión computacional**.  
- **Ejemplos:** correos electrónicos, comentarios en redes sociales, imágenes médicas, grabaciones de audio, videos.  

📌 **Ejemplo:**  
*"El servicio fue excelente, volveré a comprar"* (comentario de cliente).  

---

## 🔹 **4. Datos tipo Record**  

- Se basan en **registros independientes**, que pueden tener atributos variados.  
- Suelen encontrarse en **logs de sistemas, transacciones o datos de sensores**.  
- Mezclan características de **datos estructurados y semi-estructurados**.  
- **Ejemplos de formatos:** `.parquet`, `.avro`, archivos de logs, bases **NoSQL**.
